## Q5. For which courses the percentage of male/female students is over 70%?

```python
df_1 = pd.merge(df_students, df_classrooms, on='StudentId')
df_2 = pd.merge(df_1, df_courses, on='CourseId')

df_2.groupby('CourseId')['Gender'].value_counts()
df_3 = df_2.groupby('CourseId')['Gender'].value_counts(normalize=True).reset_index(name='counts')
df_3['counts'] = df_3['counts'].astype(float)
df_4 = df_3[df_3['counts'] > 0.7]
df_5 = pd.merge(df_4, df_courses, on='CourseId')
final = df_5[['CourseId', 'CourseName', 'Gender', 'counts']]

for i in final.index:
    final.at[i, "counts"] *= 100
    
final
```

## Q6. For each department, how many students passed with a grades over 80?

```python
df_Class_Cour = pd.merge(df_classrooms, df_courses, how='inner', left_on='CourseId', right_on='CourseId')
total_student = df_Class_Cour.groupby(['DepartmentID']).nunique().iloc[:, 0:2]
total_student = total_student["StudentId"]

df_Class_Cour = df_Class_Cour[df_Class_Cour['degree'] > 80.0] 
sudent_over_80 = df_Class_Cour.groupby(['DepartmentID']).nunique()
sudent_over_80 = sudent_over_80["StudentId"]

df_6 = pd.merge(df_deparntments, sudent_over_80, how='inner', left_on='DepartmentID', right_on='DepartmentID')
df_6 = pd.merge(df_6, total_student, how='inner', left_on='DepartmentID', right_on='DepartmentID')
df_6['Student_80_pct'] = (df_6['StudentId_x'] / df_6['StudentId_y'])*100.0
df_6.columns = ['Department_Id ','Department_Name','Student_Count Over 80','Tot_Student','Student_Percent Over 80'] 
print(df_6.to_string(index=False))
```

## Q7. For each department, how many students passed with a grades under 60?
